INSERT DATA INTO MONGODB

CONVERT TO DOCUMENTS

```python
app_docs = app_events.to_dict(
"records"
)

incident_docs = incidents.to_dict(
"records"
)

vehicle_docs = vehicles.to_dict(
"records"
)

hub_docs = hubs.to_dict(
"records"
)

delivery_docs = deliveries.to_dict(
"records"
)

complaint_docs = complaints.to_dict(
"records"
)
```

Insert collection

```python
db.app_events.insert_many(app_docs)

db.incidents.insert_many(incident_docs)

db.vehicles.insert_many(vehicle_docs)

db.hubs.insert_many(hub_docs)

db.deliveries.insert_many(delivery_docs)

db.complaints.insert_many(complaint_docs)
```

AGGREGATION ANALYSIS

APP EVENT ANALYSIS

```python
pipeline = [

{
"$group":
{
"_id":"$event_type",
"count":{"$sum":1}
}
}

]

list(db.app_events.aggregate(pipeline))
```

INCIDENT ANALYSIS

```python
pipeline = [

{
"$group":
{
"_id":"$incident_type",
"count":{"$sum":1}
}
}

]

list(db.incidents.aggregate(pipeline))
```

HUB PERFORMANCE ANALYSIS

```python
pipeline = [

{
"$group":
{
"_id":"$hub_id",
"avg_utilisation":
{
"$avg":"$utilisation_rate"
}
}
}

]

list(db.hubs.aggregate(pipeline))
```

VEHICLE ANALYSIS

```python
pipeline = [

{
"$group":
{
"_id":"$status",
"count":{"$sum":1}
}
}

]

list(db.vehicles.aggregate(pipeline))
```

DELIVERY FAILURE ANALYSIS

```python
pipeline = [

{
"$group":
{
"_id":"$delivery_status",
"count":{"$sum":1}
}
}

]

list(db.deliveries.aggregate(pipeline))
```
