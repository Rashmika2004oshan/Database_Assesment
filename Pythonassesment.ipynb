CHECK MISSING VALUES FOR ALL TABLES

```python
print("Orders Missing Values")
print(orders.isnull().sum())

print("\nDeliveries Missing Values")
print(deliveries.isnull().sum())

print("\nCustomers Missing Values")
print(customers.isnull().sum())

print("\nComplaints Missing Values")
print(complaints.isnull().sum())

print("\nDrivers Missing Values")
print(drivers.isnull().sum())
```

CLEAN DATA (REMOVE NULL VALUES)

```python
orders = orders.dropna()

deliveries = deliveries.dropna()

customers = customers.dropna()

complaints = complaints.dropna()

drivers = drivers.dropna()
```

check after cleaning

```python
print("Orders Missing Values")
print(orders.isnull().sum())

print("\nDeliveries Missing Values")
print(deliveries.isnull().sum())

print("\nCustomers Missing Values")
print(customers.isnull().sum())

print("\nComplaints Missing Values")
print(complaints.isnull().sum())

print("\nDrivers Missing Values")
print(drivers.isnull().sum())
```

MERGE 4+ TABLES

```python
merged1 = pd.merge(
    orders,
    deliveries,
    on="order_id"
)
```

```python
merged2 = pd.merge(
    merged1,
    customers,
    on="customer_id"
)
```

```python
full_data = pd.merge(
    merged2,
    drivers,
    on="driver_id"
)
```

```python
full_data.head()
```

Merge complaints separately.

```python
complaint_data = pd.merge(
    complaints,
    customers,
    on="customer_id"
)

complaint_data.head()
```

VISUALIZATION SECTION

DELIVERY STATUS GRAPH

```python
full_data["delivery_status"].value_counts().plot(kind="bar")

plt.title("Delivery Status Distribution")

plt.xlabel("Delivery Status")

plt.ylabel("Count")

plt.show()
```

FUEL COST BY HUB GRAPH

```python
full_data.groupby("hub_id")[
    "fuel_or_charge_cost"
].mean().plot(kind="bar")

plt.title("Average Fuel Cost by Hub")

plt.xlabel("Hub ID")

plt.ylabel("Fuel Cost")

plt.show()
```

ROUTE DISTANCE HISTOGRAM

```python
full_data["route_distance_km"].plot(kind="hist")

plt.title("Route Distance Distribution")

plt.xlabel("Distance")

plt.show()
```
